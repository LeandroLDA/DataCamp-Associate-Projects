### O Dataset: Video Game Sales

**Onde baixar:** [Kaggle - Video Game Sales](https://www.kaggle.com/gregorut/videogamesales)</br></br>


**Estrutura do Arquivo (`vgsales.csv`):**</br>
*   `Rank`: Ranking de vendas globais
*   `Name`: Nome do jogo
*   `Platform`: Plataforma (Wii, NES, PS4, etc)
*   `Year`: Ano de lançamento
*   `Genre`: Gênero (Action, Sports, etc)
*   `Publisher`: Editora (Nintendo, EA, etc)
*   `NA_Sales`: Vendas na América do Norte (em milhões)
*   `EU_Sales`: Vendas na Europa (em milhões)
*   `JP_Sales`: Vendas no Japão (em milhões)
*   `Other_Sales`: Vendas no resto do mundo
*   `Global_Sales`: Total de vendas mundiais

### Briefing

Você é um cientista do mercado de jogos. Precisa entregar um relatório sobre quais plataformas, gêneros e jogos dominaram o mercado histórico.

```python
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("gregorut/videogamesales")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/vgsales.csv")
```

```python
df.info()
```

```python
df.sample(5)
```

#### Perguntas:

1.  **Qual é a editora (Publisher) que mais publicou jogos na história?**
    *   Estamos falando de quantidade de títulos lançados, não de vendas.
    *   Armazene o nome numa string: `top_publisher`.



```python
top_publisher = df["Publisher"].value_counts().idxmax()

# top_publisher = None
print(top_publisher)
```

2.  **Qual a plataforma (Platform) que gerou mais dinheiro (Global_Sales) acumulado na história?**
    *   Aqui não importa quantos jogos a plataforma tem, mas a soma total da coluna `Global_Sales` para aquela plataforma.
    *   Armazene o nome da plataforma numa string: `best_selling_platform`.



```python
best_selling_platform = df.groupby("Platform")["Global_Sales"].sum().idxmax()
# best_selling_platform = None
print(best_selling_platform)
```

3.  **Qual é o jogo do gênero "Shooter" mais vendido especificamente na Europa (`EU_Sales`)?**
    *   Filtre apenas os jogos de "Shooter". Descubra qual teve a maior venda na coluna `EU_Sales`.
    *   Armazene o **nome do jogo** (string) em `top_shooter_eu_name` e o **valor de vendas** (float) em `top_shooter_eu_sales`.



```python
# top_shooter_eu_sales = df[df["Genre"] == "Shooter"]["EU_Sales"].value_counts().idxmax()
Shooters = df[df["Genre"] == "Shooter"]

top_shooter_eu_name = Shooters["EU_Sales"].idxmax()
top_shooter_eu_name = df.loc[top_shooter_eu_name]["Name"]
top_shooter_eu_sales = Shooters["EU_Sales"].max()

print(top_shooter_eu_name)
print(top_shooter_eu_sales)
```

4.  **Em qual ano houve o maior número de lançamentos de jogos?**
    *   Considere apenas a quantidade de jogos lançados por ano.
    *   **Desafio:** A coluna `Year` pode ter valores nulos (`NaN`) e vir como float (ex: `2008.0`). Trate isso para responder com um número inteiro (ex: `2008`).
    *   Armazene o ano como inteiro (int) na variável `busiest_year`.

```python
df["Year"] = df["Year"].fillna(0)
df["Year"] = df["Year"].astype('int64')

busiest_year = df["Year"].value_counts()
busiest_year = busiest_year.idxmax()
print(busiest_year)
```

5.  **Quais jogos da "Nintendo" venderam mais de 30 milhões de cópias globais (`Global_Sales`)?**
    *   Filtre pela Publisher "Nintendo" E vendas globais acima de 30.
    *   Armazene os **nomes** desses jogos numa lista chamada `nintendo_hits`.


```python
nintendo_hits = (df["Publisher"] == "Nintendo") & (df["Global_Sales"] >= 30)

nintendo_hits = df[nintendo_hits]["Name"].tolist()

print(nintendo_hits)
```
