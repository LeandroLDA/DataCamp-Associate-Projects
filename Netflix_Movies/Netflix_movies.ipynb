````python
# Qual o país mais frequente e o tipo de conteúdo mais frequente?
# Desconsidere linhas onde o país é nulo ou vazio para esta contagem.
# Armazene as respostas em duas variáveis string: top_country e top_type.

top_country = df["country"].value_counts().idxmax()
top_type = df["type"].value_counts().idxmax()

print(top_country)
# print("-"*80)
print(top_type)
````

````python
# Qual ano de lançamento (release_year) teve a maior proporção de filmes ("Movie") em relação ao total de lançamentos daquele ano?
# Atenção: Estamos falando de proporção (ratio). Se em 2020 saíram 100 obras e 90 foram filmes, a proporção é 0.9.
# Armazene o ano como um inteiro na variável top_movie_year.
df["movie_prop_year"] = (df["type"] == "Movie")
top_movie_year = df.groupby(["release_year"])["movie_prop_year"].mean().idxmax()
print(top_movie_year)
````

````python
# Qual a classificação indicativa (rating) que possui o maior número de obras lançadas especificamente pelos "United States"?
# Filtre os dados apenas para o país "United States". Descubra qual rating (ex: TV-MA, R, PG-13) aparece mais vezes.
# Armazene a resposta como um dicionário chamado usa_top_rating_dict onde a chave é o nome da classificação e o valor é a quantidade absoluta de obras.
# Exemplo esperado: {'TV-MA': 1234} (números fictícios).
rating = df[df["country"] == "United States"]["rating"].value_counts().idxmax()
rating_counts = df[df["country"] == "United States"]["rating"].value_counts().max()

usa_top_rating_dict = {rating:rating_counts}
print(usa_top_rating_dict)
````

````python
# Qual é o título do filme (apenas "Movie") mais antigo do dataset e qual a sua duração? Encontre o filme com o menor release_year.
# Armazene o título na variável string oldest_movie_title e a duração na variável string oldest_movie_duration.
filme_ano = df[df["type"] == "Movie"]["release_year"].idxmin()

oldest_movie_title = df.loc[filme_ano]["title"]
print(oldest_movie_title)

oldest_movie_duration = str(df.loc[filme_ano]["duration"])
print(oldest_movie_duration)
````

````python
# Quais diretores têm 10 ou mais títulos no catálogo? A coluna director contém alguns valores nulos, ignore-os ou trate-os.
# Crie uma lista chamada prolific_directors contendo apenas os nomes dos diretores que aparecem 10 vezes ou mais na base de dados.

prolific_directors = df["director"].value_counts()
prolific_directors = prolific_directors[prolific_directors >= 10].index.tolist()

print(prolific_directors)
````
